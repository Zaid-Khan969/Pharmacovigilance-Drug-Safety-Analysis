**Data Cleaning & Preprocessing**

```python
# Fill missing words with "Unknown"
med = med.fillna("Unknown")

med = med.drop_duplicates()
```

```python
#Standardize Drug Names
med['name'] = med['name'].str.strip().str.lower()

# Example mapping
drug_map = {"pcm": "acetaminophen", "paracetamol": "acetaminophen"}
med['name'] = med['name'].replace(drug_map)
```

```python
text_cols = ['Chemical Class', 'Therapeutic Class', 'Action Class']
for col in text_cols:
    med[col] = med[col].str.strip().str.lower()
```

```python
side_effect_cols = [c for c in med.columns if "sideEffect" in c]

med_long = med.melt(
    id_vars=['id', 'name'],
    value_vars=side_effect_cols,
    var_name='side_effect_number',
    value_name='side_effect'
)

# Remove "Unknown" side effects
med_long = med_long[med_long['side_effect'] != "Unknown"]
```

```python
med_long.to_csv("cleaned_medicine.csv", index=False)
```

**Advanced EDA & Trend Analysis**

Not just count values, but find patterns & relationships in the dataset.
We’ll answer questions like:

Which drugs cause the widest variety of side effects?

Which side effects occur together most often?

What are the strongest drug–side effect links?

```python
from wordcloud import WordCloud
```

```python
drug_side_counts = med_cleaned.groupby("name")["side_effect"].nunique().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=drug_side_counts.values, y=drug_side_counts.index, palette="Reds_r")
plt.title("Top 10 Drugs with Most Unique Side Effects")
plt.xlabel("Number of Unique Side Effects")
plt.ylabel("Drug")
plt.show()
```

```python
# Create co-occurrence matrix
co_matrix = pd.crosstab(med_cleaned['id'], med_cleaned['side_effect'])
co_matrix = co_matrix.T.dot(co_matrix)  # multiply to get co-occurrence counts

plt.figure(figsize=(12,8))
sns.heatmap(co_matrix.iloc[:20, :20], cmap="Reds")  # just first 20 side effects
plt.title("Side Effect Co-occurrence Heatmap (Top 20)")
plt.show()
```

```python
drug_side_link = med_cleaned.groupby(["name","side_effect"]).size().reset_index(name="count")
top_links = drug_side_link.sort_values("count", ascending=False).head(20)

plt.figure(figsize=(12,6))
sns.barplot(data=top_links, x="count", y="side_effect", hue="name", dodge=False)
plt.title("Top Drug–Side Effect Links")
plt.show()
```

```python
from wordcloud import WordCloud
# Word cloud for side effects
side_text = " ".join(med_cleaned['side_effect'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(side_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Side Effects")
plt.show()

# Word cloud for drug names
drug_text = " ".join(med_cleaned['name'].dropna())
wordcloud_drug = WordCloud(width=800, height=400, background_color="white").generate(drug_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_drug, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Drugs")
plt.show()
```
